```python
import random
import matplotlib.pyplot as plt

def hat_game_simulation(n, trials=50):
    
    final_numbers = {}

    for _ in range(trials):
        numbers = list(range(1, n + 1))
        while len(numbers) > 1:
            # Randomly select two different numbers
            a, b = random.sample(numbers, 2)
            numbers.remove(a)
            numbers.remove(b)
            # Add their difference back to the list
            numbers.append(abs(a - b))

        # The last remaining number is the final number
        final_number = numbers[0]
        if final_number not in final_numbers:
            final_numbers[final_number] = 0
        final_numbers[final_number] += 1

    return final_numbers

# Simulate the hat game for n = 3 to 8, each run 50 times
results = {n: hat_game_simulation(n) for n in range(3, 9)}

# Plotting
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Probability of Getting Each Final Hat Number')

for i, n in enumerate(range(3, 9)):
    ax = axs[i//3, i%3]
    data = results[n]
    probabilities = [data[key]/50 for key in sorted(data)]
    ax.bar(sorted(data), probabilities, color='skyblue')
    ax.set_title(f'n = {n}')
    ax.set_xlabel('Final Hat Number')
    ax.set_ylabel('Probability')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Adjusting the simulation for n = 13 to 20, each run 300 times
results_expanded = {n: hat_game_simulation(n, trials=300) for n in range(13, 21)}

# Plotting for the expanded range of n
fig, axs = plt.subplots(4, 2, figsize=(15, 20))
fig.suptitle('Probability of Getting Each Final Hat Number for n = 13 to 20')

for i, n in enumerate(range(13, 21)):
    ax = axs[i//2, i%2]
    data = results_expanded[n]
    probabilities = [data[key]/300 for key in sorted(data)]
    ax.bar(sorted(data), probabilities, color='skyblue')
    ax.set_title(f'n = {n}')
    ax.set_xlabel('Final Hat Number')
    ax.set_ylabel('Probability')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
